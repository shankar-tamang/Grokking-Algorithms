2.1 Suppose you're building an app to keep track of your finances.
1. Grocecerries, 2. Movie, 3. SFBC Membership

Every day, you write down everything you spent money on. At the
end of the month, you review your expenses and sum up how much
you spent. So, you have lots of inserts and a few reads. Should you
use an array or a list? 


Ans. Array is better because we have lot of appends and few reads which is efficient and simple with array. With linkedlist the overhead of extra memory for pointers is too much without any usecase where linked list excel like random insertion in the middle. Also it has worse cache locality and slower iteration.
