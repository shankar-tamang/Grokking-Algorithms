## Problem 1.1
Q.1.1 Suppose you have a sorted list of 128 names, and you are searching thorugh it using binary serach. What's the 
maximum number of steps it would take?

Ans. A binary search takes logn for worst possible case so log(128) would be 7, so in seven steps

```python
# !pip install faker 
```

```python
from faker import Faker

fake = Faker()

names = sorted({fake.name() for _ in range(300)})

names_128 = names[:128]

for name in names_128:
    print(name)
```

```python
last_name = names_128[-1]
print(last_name)
```

```python
def binary_search(name_list, item):
    low = 0
    high = len(name_list) - 1
    steps = 0

    while low <= high:

        steps += 1
        mid = round((low + high) / 2)
        guess = name_list[mid]

        if guess == item:
            return mid, steps
        
        elif guess > item:
            high = mid - 1

        else:
            low = mid + 1

    return None
```

```python
binary_search(names_128, names_128[-1])
```
