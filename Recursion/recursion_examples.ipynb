#### If we have a bunch of box and we have to find a way to build a algorithm to look thorugh all the box. We have two approaches.

**Approach 1**


1. Make a pile of boxes to look through
2. Grab a box, and look through it.
3. If you find a box, add it to the pile to look through later
4. If you find a key, you're done!
5. Repeat


**Approach 2**


1. Look through the box.
2. If you find a box, go to step 1.
3. If you find a key, you're done.


### Implementations

**Approach 1**

```python
def look_for_key(main_box):
    pile = main_box.make_a_pile_to_look_through()
    while pile is not empty:
        box = pile.grab_a_box()
        for item in box:
            if item.is_a_box():
                pile.append(item)
            elif item.is_a_key():
                print "found the key!"

```


**Approach 2**

```python
def look_for_key(box):
    for item in box:
        if item.is_a_box();
            look_for_key(item)
        elif item.is_a_key():
            print "found the key!"

```

### Base case and recursive case

Base case are very important because recursive function calls itself and it's easy to write a function incorrectly that ends up in an infinite loop.

```python
def countdown(i):
    print i
    countdown(i-1)

```

This will run forever......


Every recursive function has two parts: the base case, and the recursive case.

- Recursive case: when the function calls itself
- base case: when the function doesn't call itself again, so it doesn't go into an infinite loop


with a base case:

```python

def countdown(i):
    print i
    if i <= 0:    <-------- Base case
        return
    else:         <-------- Recursive case
        countdown(i - 1)

```


````python
def count(n):

    print(n)

    if n <= 1:
        return
    
    else:
        count(n-1)
````
