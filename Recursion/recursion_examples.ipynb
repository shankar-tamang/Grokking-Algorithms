#### If we have a bunch of box and we have to find a way to build a algorithm to look thorugh all the box. We have two approaches.

**Approach 1**


1. Make a pile of boxes to look through
2. Grab a box, and look through it.
3. If you find a box, add it to the pile to look through later
4. If you find a key, you're done!
5. Repeat


**Approach 2**


1. Look through the box.
2. If you find a box, go to step 1.
3. If you find a key, you're done.

### Base case and recursive case

Base case are very important because recursive function calls itself and it's easy to write a function incorrectly that ends up in an infinite loop.

```python
def countdown(i):
    print i
    countdown(i-1)

```

This will run forever......


Every recursive function has two parts: the base case, and the recursive case.

- Recursive case: when the function calls itself
- base case: when the function doesn't call itself again, so it doesn't go into an infinite loop


with a base case:

```python

def countdown(i):
    print i
    if i <= 0:    <-------- Base case
        return
    else:         <-------- Recursive case
        countdown(i - 1)

```


````python
def count(n):

    print(n)

    if n <= 1:
        return
    
    else:
        count(n-1)
````

## **The stack**

staStack is a simple data structure which has two functions.
- *push* (insert)
- *pop* (remove and read)





- Every time you make a function call, your computer saves the values for all the variables for that call in memory like this.
- When another funtion is called inside that function, again computer allocates a box of memory for this function call.










- When we don't create any new function or we arriave in base case, the top of the stack gets popped off. Then we return back to the function that are down.
- When you call a fucntion from another function, the calling function is paused in a partially completed state.
- You picked up after the function returns and complete the function at the last.

**3.2** Suppose you accidentally write a recursive function that runs forever. As you saw, your computer allocates memory on the stack for each function call. What happens to the stack when your recursive function runs forever.

**Ans** Ech recursive function call uses a portion of the call stack to store information like local variables, return addresses, and function context. If the recursion never ends:
1. The stack keeps growing with each new call.
2. Eventually, the stack reaches its maximum size limit (this limit is set by the operating system or the programming language runtime).
3. At that point, the program throws a stack overflow error (or RecursionError in Python).

If the system has no safegaurds or memory limits, the program could consume all available memory, potentially cuasing the device to crash.
