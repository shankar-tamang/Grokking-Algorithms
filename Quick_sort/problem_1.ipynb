**4.1** Write out the code for the earlier sum function.

```python
def total_sum(arr):

    
    if len(arr) == 1:
        return arr[0]
    
    else:
        return arr[0] + total_sum(arr[1:])
    
total_sum([5, 3, 2, 1])
```

**Note**: For a empty list, we'll get error if we try to access any value via indexing but won't get error if we are doing slicing.

**4.2** Write a recursive function to count the number of items in a list.


```python
def count_list(arr):

    if not arr:
        return 0
    
    return 1 + count_list(arr[1:])
```

```python
count_list([1, 2, 3])
```

**4.3** Find the maximum number in a list.

```python
def max_num(arr):

    if not arr:
        return float("-inf") # this is very important for empty case and we can't return 0 because it will fail when there will be negative numbers
    
    if arr[1:] == []:
        return arr[0]
    
    compare_val = max_num(arr[1:])

    if arr[0] > compare_val:
        return arr[0]
    
    else:
        return compare_val
    

max_num([1, -2, 3, 6])

```

**Note**: This is very important for empty case and we can't return 0 because it will fail when there will be negative numbers. Also len or max can be used here but isn't the most primitive approach, though len() is good enough. Also we call this primitive approach turing-machine-level-thinking.
