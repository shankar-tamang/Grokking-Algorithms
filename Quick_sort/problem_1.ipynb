**4.1** Write out the code for the earlier sum function.

````python
def total_sum(arr):

    if len(arr) == 1:
        return arr[0]
    
    else:
        return arr[0] + total_sum(arr[1:])
    
total_sum([5, 3, 2, 1])
````

**Note:** The above approach works but isn't the most optimal due to:
- Slicing: 
    - Since slicing the array creates new copies of the array. That means each `arr[1:]` allocates new memory and copies elements.
    ```python
        Call 1 → arr = [1,2,3,4]
        Call 2 → arr = [2,3,4]
        Call 3 → arr = [3,4]
        Call 4 → arr = [4]
        Call 5 → arr = []
    ```
    Slicing cost:
    ` arr[1:] on length n → copies n-1 elements `

    Total work:
    `(n-1) + (n-2) + (n-3) + ... + 1`

    Which equals: 
         O(n<sup>2</sup>) 

#### Another detail to add is that in each recursion call the array aren't copies instead they are referenced so index based recursion works efficiently for us

````python
def f(arr, count):

    if count > 0:
        count  -= 1
        f(arr, count)
        print(id(arr))
        count = count 

f([1, 2], 2) # Same address will be printed
````

#### **Index Version**

In this version we use a index to track the next adding element of the array until we get to n(total number of elements), then the base case will activate and return 0. Here index tracking can be use to understand or dictate the flow, we start from 0 to n, where 0 to n-1 is for recursive case and last n is for base case.

````python
def total_sum(arr, i=0):

    if i == len(arr):
        return 0
    
    return arr[i] + total_sum(arr, i+1)

total_sum([1, 2, 3, 4, 5])
````

**Note**: For a empty list, we'll get error if we try to access any value via indexing but won't get error if we are doing slicing.

````python
def count_list(arr):

    if not arr:
        return 0
    
    return 1 + count_list(arr[1:])
````

````python
def count_list(arr, i = 0):
    if i == len(arr):
        return 0
    
    return 1 + count_list(arr, i+1)
````

````python
def count_list(arr):
    try:
        next(arr)
        return 1 + count_list(arr)
    except StopIteration:
        return 0
    
    
````

**4.3** Find the maximum number in a list.

````python
def max_num(arr):

    if not arr:
        return float("-inf") 
    if arr[1:] == []:
        return arr[0]
    
    compare_val = max_num(arr[1:])

    if arr[0] > compare_val:
        return arr[0]
    
    else:
        return compare_val
    

max_num([1, -2, 3, 6])

````

````python
def find_max(arr):

    if not arr:
        return float("-inf")
    
    if not arr[1:]:
        return arr[0] 
    
    guess_max = find_max(arr[1:])

    if arr[0] > guess_max:
        return arr[0]
    else:
        return guess_max

find_max([-2, -3])
````

**Note**: This is very important for empty case and we can't return 0 because it will fail when there will be negative numbers. Also len or max can be used here but isn't the most primitive approach, though len() is good enough. Also we call this primitive approach turing-machine-level-thinking.


````python
def find_max(arr, i=0):

    if not arr:
        return None
    
    if i == len(arr):
        return float("-inf") # here we are comparing not doing arithmetic operation so -inf for the base case is fine
    
    compare_val = find_max(arr, i+1)

    # return arr[i] if arr[i] > compare_val else compare_val
    return max(arr[i], find_max(arr, i+1))

find_max([-1, -2, -3])
````

**4.4** Remember binary search from chapter 1? It’s a divide-and-conquer algorithm, too. Can you come up with the base case and recursive case for binary search?

````python
# Simple binary search

def simple_binary_search(arr, item):
    high = len(arr) - 1
    low = 0

    while low <= high:
        mid = round((high + low)/2)
        guess = arr[mid]

        if guess == item:
            return mid
        
        elif guess <= item:
            low = mid + 1

        else:
            high = mid - 1

    return None
````

````python
def binary_search(arr, item):

    high = len(arr) - 1
    low = 0

    mid = round((high + low)/2)
    guess = arr[mid]

    if item == guess:
        return mid
    
    elif guess > item:
        return binary_search(arr[low:mid], item)

    else:
        return binary_search(arr[mid+1:high+1], item)

binary_search([1, 2, 3, 4, 5], 5)
    
````

**Note**: In the above approach there are few problems:
1. Base case for "not found" is missing
    - if the list becomes empty, recursion never stop properly
    
2. Index is lost when slicing
    - If the array are slices, indicies restart from 0, so returning mid is no longer correct unless you track `offsets`.

````python
def binary_search(arr, item, low, high):

    if low > high:
        return None
    
    mid = (low + (high - low)//2)

    if arr[mid] == item:
        return mid
    
    elif item < arr[mid]:
        return binary_search(arr, item, low, mid-1)
    
    else:
        return binary_search(arr, item, mid+1, high)
    
binary_search([1,2,3,4,5], 4, 0, 4)
````

````python
def binary_search(arr, item):
    return _binary_search(arr, item, 0, len(arr)-1)

def _binary_search(arr, item, low, high):

    if low > high:
        return None
    
    mid = low + (high - low) // 2

    if arr[mid] == item:
        return mid
    
    elif item < arr[mid]:
        return _binary_search(arr, item, low, mid-1)
    
    else:
        return _binary_search(arr, item, mid+1, high)
````

**Note**: The above one is the standard implementation of binary_search using divide and conquer. Using offset and using slicing array is useful but has bad space complexity.

The above implementation has few important part:
- low > high this is very important for when we don't have the search item and also handles for empty list condition. The second approach of binary_search is the standard way to use in system or libraries.
- mid = (low + (high - low)//2) , this is a approach to avoid bit overflow.
- We use floor division because it is safer and compared to round which uses banker's rounding it is more deterministic.
- Parameter size of this approach is greater but is efficient.

**Tips**: For divide and conquer algorithm using recursive and array, we don't have to pass a copy of the arr by slicing, we just pass indexes as a counter or a tracker.

-  This is called `Divide & Conquer` with Range Parameters.
    ` array + start index + end index `
- Also called Structural Recursion on Intervals
    **we decompose**
    `problem(l, r)`

    **into**
    ```
    problem(l, mid)
    problem(mid+1, r)
    
    ```




### **QUICK SORT**

````python
def quicksort(array):
    if len(array) < 2:
        return array 
    
    else:
        pivot = array[0]
        less = [i for i in array[1:] if i <= pivot]

        greater = [i for i in array[1:] if i >= pivot]

        return quicksort(less) + [pivot] + quicksort(greater)
    
quicksort([10, 5, 2, 3])
````

**Note**: 
- in the base case, we need it to < or > because we have to deal with single value list and empty list as well, for empty it will return the empty list
- array should be returned instead of teh values, otherwise it will cause mismatch operation of addition between list and int

**Tips:**
- For each type of case, base case's logic is very important, and for array's related problem it should make up for empty and list with only one element.

| Operation | Base Case Return |
| --------- | ---------------- |
| Sum       | `0`              |
| Product   | `1`              |
| Max       | `-inf`           |
| Min       | `+inf`           |
